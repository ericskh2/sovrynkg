```python
#export

import pandas as pd
import numpy as np
from pydantic import BaseModel, validator

def short(address):
    return address[:5]+'â€¦'+address[-4:]

class Address(BaseModel):
    objtype:str
    prefix:str
    name: str
    address:str
        
    @validator('prefix')
    def prefix_is_not_nan(cls, prefix_str):
        if prefix_str == 'nan':
            return ''
        return prefix_str
    @validator('address')
    def address_is_lower(cls, address_str):
        return address_str.lower()
    def desc(self):
        if self.prefix:
            return '{pre}_{n}'.format(pre=self.prefix, n=self.name)
        else:
            return self.name
    
    def __repr__(self):
        return str(self)
    def __str__(self):
        return '<{objtype} {d}:{sa}>'.format(objtype=self.objtype, 
                                             d=self.desc(), 
                                             sa=short(self.address))

class Token(Address):
    pass

class Contract(Address):
    pass

OBJTYPES = {'TOKEN': Token,
           'CONTRACT': Contract,
            'ADDRESS': Address
           }
```
